```python
## transfomration of to a date time object
df['Date Joined'] = pd.to_datetime(df['Date Joined'])
df.dtypes
```

```python
## Create a month of joined column
## using the function month_name from the datetime module

df['Month Joined'] = df['Date Joined'].dt.month_name()
df.head()
```

```python
## Create anothe column with the difference in years from created to joined

df['Year Joined'] = df['Date Joined'].dt.year - df['Year Founded']
df.head()
```

```python
## Selecting years to observe trends.
## Creating groups based on specific periods
## of time can help us understand behaiviours
## that may be in cycles.

df15 = df[df['Date Joined'].dt.year == 2015]
df15 = df15.copy()

df16 = df[df['Date Joined'].dt.year == 2016]
df16 = df16.copy()
```

```python
## Looking for trends in the year 2015
## Here we are looking into the number of week in the year.
## the  format goes yyyy-W[1-56]

df15["Week Joined"] = df15["Date Joined"].dt.strftime('%Y-W%V')
df15_by_week = df15.groupby(by="Week Joined")["Company"].count().reset_index().rename(columns={"Company":"Company Count"})

## Looking for trends in the year 2016
df16["Week Joined"] = df16["Date Joined"].dt.strftime('%Y-W%V')
df16_by_week = df16.groupby(by="Week Joined")["Company"].count().reset_index().rename(columns={"Company":"Company Count"})
```

**How to structure the data to observe trends over time?**
lots of ways.

- One approach may consists of the following:

  - Identify 2 things to compare.
  - Choose a time interval over which you want to observe average valuation.
  - Create a subset of data that corresponds to the additional year. Concatenate that subset with the previous subset you created, in order to get the data for both years in one DataFrame. 
  - Create a column that contains the time interval that each data point belongs to (as needed), group by that column, and compute the average funding of companies that joined per interval. 
  
- For example, to compare trends in average funding of companies quarterly between 2015 and 2016, if you already had a subset for 2016, you could do the following: 
- Create a subset for 2015
- Concatenate that with the subset for 2016.
- Create a column that corresponds to quarter joined.
- Group by quarter.
- Compute the average valuation of companies that joined per quarter. 

This allows the observation of trends over the quarters of 2015 compared to 2016 


```python
## Concatenate that with the subset for 2016.
companies_15_16 = pd.concat([df15, df16])
```

```python
## Create a column that corresponds to quarter joined.
companies_15_16['Quarter Joined'] = companies_15_16['Date Joined'].dt.to_period('Q').dt.strftime('%Y-Q%q') 
companies_15_16.head()
```

```python
## convert column valuation to numeric and take the B out
companies_15_16['AVG Valuation'] = companies_15_16['Valuation'].apply(lambda j: j.strip("$B")).astype(float)
```

```python
## Compute the average valuation of companies that joined per quarter.
companies_15_16_byqt = companies_15_16.groupby(by='Quarter Joined')['AVG Valuation'].mean().reset_index()
companies_15_16_byqt
```

# Statistical test

Looking into how long it took the companies to join the unicorn group with respect to the month they joined.
the moths will go chronological using a list of ordered months.
Using all the data.

```python
months = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", 
               "December"]

plt.figure(figsize = (15, 7))
sns.boxplot(
    x = df['Month Joined'], 
    y = df['Year Joined'],
    order = months,
    showfliers=False,
)
## Setting title

plt.title('Distribution of the years it took to become unicorn company with respect to the month joined')
## Rotate labels
plt.xticks(rotation=45,  horizontalalignment='right')
#Display plot
plt.show()
```

### Considerations from the boxplot
---
- #### The median value for `Years To Join` is different for each month, around 4.9 and 7.3. 
- #### Also, the median `Years To Join` is lower for the months of September and October. 
- #### **This could indicate that companies that reached unicorn status in early fall took less time to reach $1 billion valuation.** 
- #### This is because the number of companies is relatively close/consistent for each month. 
- #### If that were not the case, it would be misleading to compare the median values from the box plots between months.


# The average valuation over the quarters, 

with two bars for each time interval. 
There will be two bars for each time interval. 
This allows to compare quarterly values between the two years.


```python

# Using slicing, extract the year component and the time interval that you specified, 
# and save them by adding two new columns into the subset. 

companies_15_16_byqt['Quarter Number'] = companies_15_16_byqt['Quarter Joined'].str[-2:]
companies_15_16_byqt['Year Joined'] = companies_15_16_byqt['Quarter Joined'].str[:4]

# Set the size of the plot.
plt.figure(figsize = (10, 5))
# Create a grouped bar plot.
sns.barplot(
    x = companies_15_16_byqt['Quarter Number'],
    y = companies_15_16_byqt['AVG Valuation'],
    hue = companies_15_16_byqt['Year Joined']
)
plt.plot()
# Set the x-axis label.
plt.xlabel("Quarter number")
# Set the y-axis label.
plt.ylabel("Average valuation (billions of dollars)")
# Set the title.
plt.title("Average valuation of companies that became unicorns per quarter in 2015 vs. 2016")
# Display the plot.
plt.show()
```

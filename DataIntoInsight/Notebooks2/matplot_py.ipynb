```python
# f(x) = ax + b

slope = float(input('enter the slope:'))
interceptor = float(input('enter the interceptor:'))
x = [i for i in range(-30, 30)]
y = [(slope * element + interceptor) for element in x]
c = [random.randint(-30, 30) for i in range(-30, 30)]


  
# a line 
fig5, ax5 = plt.subplots()
ax5.scatter(x, y, c=c, alpha=0.3)
ax5.set_title('a Line')
ax5.set_xlabel(" X ")
ax5.set_ylabel("f(x) = ax + b = y")
ax5.vlines(x=0, ymin=-30, ymax=30, colors='r', linestyles='dotted')
ax5.hlines(y=0, xmin=-30, xmax=30, colors='r', linestyles='dotted')
ax5.set_aspect('equal', 'box')
ax5.set(xlim=(-30, 30), ylim=(-30, 30))

```

### Horsepower vs Torque
- Import matplotlib.pyplot as plt and use %matplotlib inline.
- Import csv
- Read the files/cars.csv wil into a list of dictionaries.
- Use list comprehension to get
    - Horsepwer
    - Torque
    - Bonus: Fuel City mpg
- Make a scatter plot
    - Options: Use alpha=.25 and c= with the list of Fuel City mpg

```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
with open("/home/gato/Scripts/py_ex/LearnPython-main/files/cars.csv", "r") as f:
    csv_reader = csv.DictReader(f)
    cars = list(csv_reader)
```

```python
print(cars[0])
print(len(cars))
```

```python
horsepower = [int(keys['Horsepower']) for keys in cars]
torque = [int(keys['Torque']) for keys in cars]
fuelCityMpg = [int(keys['Fuel City mpg']) for keys in cars]
```

```python
print(max(fuelCityMpg))
print(min(fuelCityMpg))
print(sum(fuelCityMpg)/(len(fuelCityMpg)))
```
